```python
import pandas as pd
col_names = ["duration","protocol_type","service","flag","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate","label"]
```

```python
kdd_data = pd.read_csv("kddcup.data", header=None, names = col_names)
kdd_data.keys() # keys == column names
# kdd_data
```

```python
kdd_data['label'].value_counts()
```

```python
num_features = [
    "duration","src_bytes",
    "dst_bytes","land","wrong_fragment","urgent","hot","num_failed_logins",
    "logged_in","num_compromised","root_shell","su_attempted","num_root",
    "num_file_creations","num_shells","num_access_files","num_outbound_cmds",
    "is_host_login","is_guest_login","count","srv_count","serror_rate",
    "srv_serror_rate","rerror_rate","srv_rerror_rate","same_srv_rate",
    "diff_srv_rate","srv_diff_host_rate","dst_host_count","dst_host_srv_count",
    "dst_host_same_srv_rate","dst_host_diff_srv_rate","dst_host_same_src_port_rate",
    "dst_host_srv_diff_host_rate","dst_host_serror_rate","dst_host_srv_serror_rate",
    "dst_host_rerror_rate","dst_host_srv_rerror_rate"
]
features = kdd_data[num_features].astype(float)
features
```

```python
labels = kdd_data['label'].copy()
labels[labels!='normal.'] = 'attack.'
labels.value_counts()
```

```python
features.describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
features, labels, random_state=0)
```

```python
from time import time
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3, algorithm = 'ball_tree', leaf_size=500)
print(knn)
t0 = time()
knn.fit(X_train, y_train)
tt = time()-t0
print("Classifier trained in {} seconds".format(round(tt,3)))
```

```python
from sklearn.tree import DecisionTreeClassifier
clf_tree = DecisionTreeClassifier( max_depth = 5 )
t0 = time()
clf_tree.fit( X_train, y_train )
tt = time()-t0
print("Classifier trained in {} seconds".format(round(tt,3)))
```

```python
pd.DataFrame(clf_tree.predict_proba(X_train))
```

```python
tree_train_pred = pd.DataFrame( { 'actual':  y_train,
                            'predicted': clf_tree.predict( X_train ) } )
```

```python
tree_train_pred.head(100)
```

```python
tree_test_pred = pd.DataFrame( { 'actual':  y_test,
                            'predicted': clf_tree.predict( X_test ) } )
```

```python
tree_test_pred.head(10)
```

```python
from sklearn import metrics
metrics.accuracy_score( tree_train_pred.actual, tree_train_pred.predicted )
```

```python
metrics.accuracy_score( tree_test_pred.actual, tree_test_pred.predicted )
```
